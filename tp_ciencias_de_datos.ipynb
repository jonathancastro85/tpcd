## A - Comprensión del negocio (Business Understanding - CRISP-DM)

### **Contexto del problema**

Sector gastronomico, industria de restaurantes y comidas

**Por que es relevante:** En la industria gastronomica, los margenes de ganacia suelen ser ajustados. Una gestion ineficiente del stock o desconocimiento de los productos que sostienen el negocio, pueden llevar a problemas financieros a futuro. Analizar estos datos permite ayudar en la toma de decisiones basado en evidencia

**Quienes se benefician:** El dueño o gerente del restaurante, para maximizar la rentabilidad general
***El encargado del inventario***, para optimizar los pedidos a proveedores para evitar sobrestock o falta de stock
**Clientes**, para disfrutar de lo que quieran sin que haya problemas de stock

###**Analizar las ventas de un local gastronómico**

Identificar patrones de consumo, métodos de pago y posibles inconsistencias en los datos.

###**Objetivos**

Analizar evolución de ventas.

Identificar los métodos de pago más utilizados.

Detectar problemas en los datos (fechas mal formateadas, tipos de pago inconsistentes).

Preparar los datos para futuros análisis (limpieza y transformación).

Identificar productos mas utilizados.

Identificar ordenes(menues) mas populares

###**Preguntas a resolver**

¿Qué días y horarios se vende más?

¿Qué método de pago es más popular?

¿Hay errores en las fechas o registros duplicados?

¿Cómo afectan los datos sucios al análisis?

Ordenes mas populares(menues)

###**Problemas detectados**

Fechas en distintos formatos (ej: dd/mm/yyyy , mm-dd-yyyy , etc.).

Tipos de pago inconsistentes (ej: "Cash", "cash", "CASH")

Posibles valores nulos o vacíos
